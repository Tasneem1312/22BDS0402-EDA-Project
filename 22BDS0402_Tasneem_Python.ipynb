Smart Library Management System using Python Scriptin

```python
import sys

books = {}
users = {}

def initialize_system():
    """Q1: Initialize system and display welcome message."""
    print("=" * 50)
    print("Welcome to the Smart Library Management System".center(50))
    print("=" * 50)

def add_book():
    """Q2, Q3, Q5: Add books with validation and looping."""
    while True: # Q5: Loop to allow adding multiple books
        print("\n--- Add New Book ---")
        try:
            # Q3: Validate Book ID is an integer
            b_id = int(input("Enter Book ID (Integer): "))

            # Q3: Validate Book ID is unique
            if b_id in books:
                print(f"Error: Book ID {b_id} already exists!")
                continue

            title = input("Enter Book Title: ")
            author = input("Enter Author Name: ")

            # Q2: Store in data structure (Dictionary)
            books[b_id] = {
                "title": title,
                "author": author,
                "available": True
            }
            print(f"Book '{title}' added successfully.")

        except ValueError:
            print("Invalid Input! Book ID must be a number.")
            continue

        # Check if user wants to add another book
        cont = input("Add another book? (y/n): ").lower()
        if cont != 'y':
            break

def register_user():
    """Q6: Accept user details and store them."""
    print("\n--- Register User ---")
    try:
        u_id = int(input("Enter User ID (Integer): "))
        if u_id in users:
            print(f"Error: User ID {u_id} already exists!")
            return

        name = input("Enter User Name: ")

        users[u_id] = {
            "name": name,
            "borrowed_count": 0
        }
        print(f"User {name} registered successfully.")

    except ValueError:
        print("Invalid Input! User ID must be a number.")

def borrow_book_func(user_id, book_id):
    """Q10: Function to borrow book and update availability."""
    # Update book status
    books[book_id]['available'] = False
    # Update user count
    users[user_id]['borrowed_count'] += 1
    print(f"Success! '{books[book_id]['title']}' issued to {users[user_id]['name']}.")

def handle_borrowing():
    """Q4, Q8: Logic to handle input and check eligibility."""
    print("\n--- Borrow a Book ---")
    try:
        u_id = int(input("Enter User ID: "))
        b_id = int(input("Enter Book ID: "))

        # Q8: Relational and Logical operators for eligibility
        # Check if user exists
        if u_id not in users:
            print("Error: User not registered.")
            return

        # Check if book exists
        if b_id not in books:
            print("Error: Book not found.")
            return

        user = users[u_id]
        book = books[b_id]

        # Q4 & Q8: Check eligibility (Registered, Count < 3, Available)
        if (user['borrowed_count'] < 3) and (book['available'] == True):
            borrow_book_func(u_id, b_id)
        else:

            if book['available'] == False:
                print("Error: Book is currently unavailable.")
            elif user['borrowed_count'] >= 3:
                print("Error: User has reached maximum borrowing limit (3).")

    except ValueError:
        print("Invalid Input! Please enter numeric IDs.")

def return_book():
    print("\n--- Return a Book ---")
    try:
        u_id = int(input("Enter User ID: "))
        b_id = int(input("Enter Book ID: "))

        if u_id in users and b_id in books:
            if not books[b_id]['available']:
                books[b_id]['available'] = True
                if users[u_id]['borrowed_count'] > 0:
                    users[u_id]['borrowed_count'] -= 1
                print("Book returned successfully.")
            else:
                print("Error: This book was not borrowed.")
        else:
            print("Error: Invalid User ID or Book ID.")
    except ValueError:
        print("Invalid input.")

def view_available_books():
    """Q9: Use list comprehension to display available books."""
    print("\n--- Available Books ---")

    # Q9: List Comprehension
    available_list = [
        f"ID: {bid} | {details['title']} by {details['author']}"
        for bid, details in books.items()
        if details['available'] == True
    ]

    if available_list:
        for item in available_list:
            print(item)
    else:
        print("No books currently available.")

def main():
    """Q7: Menu-driven system."""
    initialize_system()

    while True:
        print("\n=== MAIN MENU ===")
        print("1. Add Book")
        print("2. Register User")
        print("3. Borrow Book")
        print("4. Return Book")
        print("5. View Available Books")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            add_book()
        elif choice == '2':
            register_user()
        elif choice == '3':
            handle_borrowing()
        elif choice == '4':
            return_book()
        elif choice == '5':
            view_available_books()
        elif choice == '6':
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice! Please select 1-6.")

if __name__ == "__main__":
    main()
```

3. Create patient.py

```python
%%writefile hospital_system/patient.py
from .utils import UnderAgeException
from .logger import log_exception # Import our new logger
import shutil # For backup

FILE_NAME = "patients.txt"

# Q4: Handle FileNotFoundError (Helper function)
def read_all_lines():
    try:
        with open(FILE_NAME, "r") as f:
            return f.readlines()
    except FileNotFoundError:
        return []

def register_patient(p_id, name, age, consult_type):
    """Q1: Write details. Q2: Append without overwriting."""
    try:
        # Validate age (Logic from Ex 2)
        if int(age) < 18:
            raise UnderAgeException("Patient is under 18.")

        # Save to file (Append mode 'a')
        with open(FILE_NAME, "a") as f:
            # Format: ID,Name,Age,Type
            record = f"{p_id},{name},{age},{consult_type}\n"
            f.write(record)

        print(f"Success: Patient {name} saved to {FILE_NAME}.")
        return True

    except UnderAgeException as e:
        print(f"Validation Error: {e}")
        log_exception(str(e)) # Q10
    except Exception as e:
        print(f"Error saving patient: {e}")
        log_exception(str(e)) # Q10

def display_all_patients():
    """Q3: Read patient data and display formatted output."""
    lines = read_all_lines()
    if not lines:
        print("No records found.")
        return

    print(f"{'ID':<10} {'Name':<15} {'Age':<5} {'Type':<15}")
    print("-" * 50)
    for line in lines:
        try:
            pid, name, age, ctype = line.strip().split(',')
            print(f"{pid:<10} {name:<15} {age:<5} {ctype:<15}")
        except ValueError:
            continue # Skip malformed lines

def search_patient_by_id(search_id):
    """Q8: Search patient details from file using Patient ID."""
    lines = read_all_lines()
    found = False
    for line in lines:
        if line.startswith(str(search_id) + ","):
            print("\n--- Patient Found ---")
            print(f"Record: {line.strip()}")
            found = True
            break
    if not found:
        print("Patient ID not found.")
    return found

def count_total_patients():
    """Q9: Count total number of patients by reading file."""
    lines = read_all_lines()
    return len(lines)

def backup_data():
    """Q11: Backup patient data to backup_patients.txt."""
    try:
        shutil.copy(FILE_NAME, "backup_patients.txt")
        print("Backup successful: Created backup_patients.txt")
    except FileNotFoundError:
        print("Error: No patient data file to backup.")
        log_exception("Backup failed: patients.txt not found")
```

4. Create appointment.py

```python
%%writefile hospital_system/appointment.py
from .logger import log_exception

FILE_NAME = "appointments.txt"

def schedule_appointment(p_id, name, consult_type):
    """Q5: Store appointment details in structured format."""
    try:
        with open(FILE_NAME, "a") as f:
            # Format: ID,Name,Type
            f.write(f"{p_id},{name},{consult_type}\n")
        print("Appointment confirmed and saved to file.")
    except Exception as e:
        print("Error scheduling appointment.")
        log_exception(f"Appointment Write Error: {e}")

def view_appointments_by_type(filter_type):
    """Q6: Read and filter appointments by consultation type."""
    try:
        with open(FILE_NAME, "r") as f:
            lines = f.readlines()

        print(f"\n--- Appointments ({filter_type}) ---")
        found = False
        for line in lines:
            parts = line.strip().split(',')
            if len(parts) >= 3:
                pid, name, ctype = parts
                if ctype.lower() == filter_type.lower():
                    print(f"ID: {pid} | Patient: {name}")
                    found = True

        if not found:
            print("No appointments found for this type.")

    except FileNotFoundError:
        print("No appointment records found.")
```

5. Create billing.py

```python
%%writefile hospital_system/billing.py
import datetime
from .logger import log_exception

BILLING_FILE = "billing.txt"
FEE_STRUCTURE = {"General": 500, "Specialist": 1000, "Emergency": 1500}

def calculate_fees(consult_type):
    return FEE_STRUCTURE.get(consult_type, 500)

def generate_bill(p_id, name, consult_type):
    """Q7: Write billing details with date and amount."""
    try:
        amount = calculate_fees(consult_type)
        tax = amount * 0.18
        total = amount + tax
        date_str = datetime.datetime.now().strftime("%Y-%m-%d")

        # Write to file
        with open(BILLING_FILE, "a") as f:
            f.write(f"{date_str},{p_id},{name},{total:.2f}\n")

        print(f"\n--- Bill Generated for {name} ---")
        print(f"Date: {date_str}")
        print(f"Total Amount: ${total:.2f} (Saved to {BILLING_FILE})")
        return total

    except Exception as e:
        print("Error generating bill.")
        log_exception(f"Billing Error: {e}")
```
